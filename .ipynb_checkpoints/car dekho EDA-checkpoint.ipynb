```python
fig, axes = plt.subplots(2, 2, figsize=(8, 9))

# Data for each subplot
subplot_data = ['transmission', 'fuel', 'seller_type', 'owner']
subplot_titles = ['Car Owner Type', 'Transmission Type', 'Fuel Type', 'Seller Type']

for i, subplot in enumerate(subplot_data):
    ax = sns.countplot(x=df[subplot], ax=axes[i // 2, i % 2])
    for label in ax.containers:
        ax.bar_label(label)
    plt.xticks(rotation=45)
    ax.set_title(subplot_titles[i])  # Set title for each subplot

plt.tight_layout()
plt.show()
```

```python
def brand(x):
    x = x.split(' ')
    return x[0]
df['brand'] = df.name.apply(brand)
```

```python
x = df['brand'].value_counts()
x.plot(kind = 'bar')
plt.title('Car brands')
plt.show()
```

```python
y = df.groupby('brand')['year'].count()
plt.figure(figsize=(10,5))
ax = y.plot(kind = 'bar')
for label in ax.containers:
        ax.bar_label(label)
plt.show()
```

```python
df.drop(columns = 'name', inplace = True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import datasets
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn import preprocessing
  
# label_encoder object knows 
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
l = ['fuel', 'seller_type','transmission','owner']
for i in l:
    df[i]= label_encoder.fit_transform(df[i])
```

```python
x = df[['year', 'km_driven', 'fuel', 'seller_type',
       'transmission', 'owner']]
y = df[['selling_price']]
```

```python
x2_train,x2_test,y2_train,y2_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg.fit(x2_train, y2_train)
```

```python
regp = reg.predict(x2_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y2_test, regp))
print("RMSE:", rmse)
```
